### Merging Dataframe

```python
df1 = pd.DataFrame({
    "name" : ["Priyam","Rohan","Shubham","Rohit","Rahul","Ramesh"],
    "marks" : [100,90,80,70,60,50]
})
df1
```

```python
df2 = pd.DataFrame({
    "name" : ["Akshay","Akal","Apa","Prio"],
    "marks" : [10,90,80,70]
})
df2
```

```python
pd.concat([df1,df2])
```

```python
df1 = pd.DataFrame({
    "name" : ["Priyam","Rohan"],
    "marks" : [100,90]
})
df1
```

```python
df2 = pd.DataFrame({
    "name" : ["Priyam","Rohan"],
    "roll" : [1,2]
})
df2
```

```python
pd.merge(df1,df2, on="name")
```
